只顯示最大可能情緒

```python
# 定義情緒預測函數
def predict_emotion_with_probabilities(face_img, model):
    # 將人臉影像預處理
    face_img = cv2.resize(face_img, (48, 48))  # 調整大小
    face_img = face_img / 255.0  # 正規化
    face_img = np.expand_dims(face_img, axis=-1)  # 增加單通道維度
    face_img = np.expand_dims(face_img, axis=0)  # 增加批次維度
    
    # 預測情緒
    predictions = model.predict(face_img)
    probabilities = predictions[0]
    return probabilities
```

```python
import cv2
import numpy as np
import tensorflow as tf

# 定義情緒標籤
emotion_labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']


model = tf.keras.models.load_model('gray_0.59.h5')
cap = cv2.VideoCapture(0)

# 設定 OpenCV 人臉檢測模型
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

while True:
    # 讀取攝影機影像
    ret, frame = cap.read()
    if not ret:
        break

    # 將影像轉為灰階
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # 偵測人臉
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)

    for (x, y, w, h) in faces:
        # 擷取人臉區域
        face = gray[y:y + h, x:x + w]
        
        # 預測人臉情緒
        probabilities = predict_emotion_with_probabilities(face, model)
        
        # 顯示所有情緒的概率
        for i, prob in enumerate(probabilities):
            text = f"{emotion_labels[i]}: {prob*100:.2f}%"
            # 在影像上顯示每個情緒的百分比，依次往下排列
            cv2.putText(frame, text, (x, y + h + 20 + i * 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1)

        # 在人臉框上繪製矩形
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)

    # 顯示影像
    cv2.imshow('Real-Time Emotion Detection', frame)

    # 按下 'q' 鍵退出
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 釋放資源
cap.release()
cv2.destroyAllWindows()
```

顯示情緒趴數

time.sleep(0.1)可調整偵測時間間距

```python
import cv2
import numpy as np
import tensorflow as tf

# 定義情緒標籤及對應顏色
emotion_labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']
emotion_colors = {
    'Angry': (0, 0, 255),       # 紅色
    'Disgust': (0, 255, 0),     # 綠色
    'Fear': (255, 0, 0),        # 藍色
    'Happy': (0, 255, 255),     # 黃色
    'Neutral': (128, 128, 128), # 灰色
    'Sad': (255, 255, 255),     # 白色
    'Surprise': (255, 0, 255)   # 紫色
}

# model = tf.keras.models.load_model('gray_0.59.h5')
model = tf.keras.models.load_model('model_1.h5')
cap = cv2.VideoCapture(0)

# 設定 OpenCV 人臉檢測模型
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')



while True:
    # 讀取攝影機影像
    ret, frame = cap.read()
    if not ret:
        break

    # 將影像轉為灰階
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # 偵測人臉
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)

    highest_prob = 0
    highest_emotion = ''
    probabilities = []

    for (x, y, w, h) in faces:
        # 擷取人臉區域
        face = gray[y:y + h, x:x + w]
        
        # 預測人臉情緒
        probs = predict_emotion_with_probabilities(face, model)
        probabilities = probs  # 更新所有情緒的概率

        # 找到最高概率的情緒
        max_prob_idx = np.argmax(probs)
        highest_prob = probs[max_prob_idx]
        highest_emotion = emotion_labels[max_prob_idx]

        # 顯示所有情緒的概率
        for i, prob in enumerate(probabilities):
            text = f"{emotion_labels[i]}: {prob * 100:.2f}%"
            # 在影像上顯示每個情緒的百分比，依次往下排列
            cv2.putText(frame, text, (x, y + h + 20 + i * 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1)

        # 繪製當前最高情緒的框
        cv2.putText(frame, f"Highest Emotion: {highest_emotion} ({highest_prob * 100:.2f}%)", 
                    (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)
        cv2.rectangle(frame, (x, y), (x + w, y + h), emotion_colors[highest_emotion], 2)

    # 顯示影像
    cv2.imshow('Real-Time Emotion Detection', frame)
    time.sleep(0.1)

    # 按下 'q' 鍵退出
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 釋放資源
cap.release()
cv2.destroyAllWindows()
```
